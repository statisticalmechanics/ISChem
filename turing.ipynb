```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
L = 60.0
N = 100
a = L / N

h = 1e-1
print(a, h)


Du = 0.02
Dv = 0.5

au = 0.08
bu = -0.08
cu = 0.04

av = 0.1
bv = 0.0
cv = -0.15

hu = h*Du/(a*a)
hv = h*Dv/(a*a)

```

```python
u = np.empty(N, float)
v = np.empty(N, float)

su = 1
uc = 0.5
qu = 10

for i in range(N):
    u[i] = qu*np.exp(-(i-N/2)**2/(2*su**2)) / np.sqrt(2*3.1415926*su**2) + uc
    
    
sv = 8
vc = 0.4
qv = 10
for i in range(N):
    v[i] = vc #qv*np.exp(-(i-N/2+10)**2/(2*sv**2)) / np.sqrt(2*3.1415926*sv**2) + vc

u_max = np.max(u)
v_max = np.max(v)

u_sum =np.sum(u)
v_sum =np.sum(v)

u_new = np.empty(N, float)
v_new = np.empty(N, float)

print(u_max, v_max)

print(u_sum, v_sum)

u is u_new
```

```python
plt.plot(list(range(N)), u, 'r-')
plt.plot(list(range(N)), v, 'b--')
plt.ylim(0.0,)
```

```python
u0 = np.copy(u)
v0 = np.copy(v)
```

```python
plt.plot(list(range(N)), (u0-np.min(u0))/(np.max(u0) - np.min(u0)))#, 'k-')
plt.plot(list(range(N)), (v0-np.min(u0))/(np.max(u0) - np.min(u0)))#, 'k-')
```

```python
t = 0.
tend = 16001

while t < tend: 
    if (int(t) % 1000 == 0) and abs(t - int(t) < 0.01):
        print(t)

    for i in range(N):
        u_new[i] = u[i] + h*( au * u[i] + bu * v[i] + cu) + hu * (u[(i+1)%N] + u[(i-1)%N] - 2*u[i])
        v_new[i] = v[i] + h*( av * u[i] + bv * v[i] + cv) + hv * (v[(i+1)%N] + v[(i-1)%N] - 2*v[i])
        
    u = np.copy(u_new)
    v = np.copy(v_new)
     
    #if(abs(t-1000.0)<h*0.1):
    if np.max(u>1e3):
        us = np.sum(u)  
        vs = np.sum(v)
        u = np.copy(u*u_sum/us)
        v = np.copy(v*u_sum/us)
    
    if(abs(t-1.0)<h*0.1):
        u1 = np.copy(u)
        v1 = np.copy(v)
        
    if(abs(t-10.0)<h*0.1):
        u2 = np.copy(u)
        v2 = np.copy(v)
    
    if(abs(t-100.0)<h*0.1):
        u3 = np.copy(u)
        v3 = np.copy(v)
        
    if(abs(t-1000.0)<h*0.1):
        u4 = np.copy(u)
        v4 = np.copy(v)
        
    if(abs(t-10000.0)<h*0.1):
        u5 = np.copy(u)
        v5 = np.copy(v)
    
    if(abs(t-15200.0)<h*0.1):
        u6 = np.copy(u)
        v6 = np.copy(v)
        
    if(abs(t-15400.0)<h*0.1):
        u7 = np.copy(u)
        v7 = np.copy(v)
        
    t += h
```
